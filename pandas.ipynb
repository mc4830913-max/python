```python
#g) All strings functions can be used to extract or modify texts in a series
ser=pd.Series(["red colour","blue","orange","engineering","    graduate","geeks for geeks"])
print(ser.str.upper())
print("_"*20)
print(ser.str.lower())
print("_"*20)
for i in ser:
  print(i,len(i))
print("_"*20)
ser = ser.str.strip()
for i in ser:
    print(i , len(i))
print("_"*20)
print(ser.astype("str"))
print("_"*20)
print(ser.str.split())
print(ser.str.split()[5])

```

```python
cricket = {'Team'   : ['India', 'India', 'Australia', 'Australia', 'SA', 'SA', 'SA', 'SA', 'NZ', 'NZ', 'NZ', 'India'],
           'Rank'   : [2, 3, 1,2, 3,4 ,1 ,1,2 , 4,1,2],
           'Year'   : [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points' : [876,801,891,815,776,784,834,824,758,691,883,782]}
df=pd.DataFrame(cricket)
df

df.groupby('Year').groups
df.groupby(['Team','Year']).get_group(('India',2017))
df.groupby('Team').sum()['Points']
# df.groupby('Team').max()['Points']
# df.groupby('Team').min()['Points']
df.groupby('Team').sum()['Points'].sort_values(ascending=True)
groups=df.groupby('Team')
groups['Points'].agg([np.mean,np.median,np.std,np.var,np.max,np.min])
df.groupby('Team').filter(lambda x:len(x)==3)  #disply number of item appear in list
df.groupby('Rank').filter(lambda x:len(x)==4)
```
