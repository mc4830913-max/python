```python
from selenium import webdriver

driver = webdriver.Chrome()

driver.get("https://www.youtube.com/@GateSmashers/videos")
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
data=[]
for sop in soup.find_all("yt-lockup-view-model"):
    title         =sop.find("h3",class_="ytLockupMetadataViewModelHeadingReset").text
    vedio_link    =sop.find("a",class_="ytLockupViewModelContentImage").get("href")
    views         =sop.find_all("span",class_="ytAttributedStringHost ytContentMetadataViewModelMetadataText ytAttributedStringWhiteSpacePreWrap ytAttributedStringLinkInheritColor")[0].text
    date_time     =sop.find_all("span",class_="ytAttributedStringHost ytContentMetadataViewModelMetadataText ytAttributedStringWhiteSpacePreWrap ytAttributedStringLinkInheritColor")[1].text
    try:
        thumnail_link=sop.find("img").get("src").split("?")[0]
    except:
        thumnail_link=None
    data.append([title,date_time,views,vedio_link,thumnail_link])
data    
```

```python
df=pd.DataFrame(data, columns=["title","date_time","views","vedio_link","thumnail_link"])
df.head(10)
```
