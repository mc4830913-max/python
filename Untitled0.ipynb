```python
from os import lchown
#concationation
print(len("b.tech"+' student'))
print()
print('_'*20)

#replication
print("MonikaChauhan "*5)
print('_'*20)

#string length
print(len('abc'))
print(len('Monika ji '))
print('_'*20)

#string slicing
print('Monika'[-6])
print('Monika'[5])
print('technology grow day by day'[11:16])
print('terminal is not running'[0:23])
print('terminal is not running'[5:])
print('terminal is not running'[:15])
print('terminal is not running'[:-15])
print('terminal is not running'[-5:])
print('_'*20)

#function of string in python
#string case conversion
print('hello everyone'.lower())
print('hello everyone'.upper())
print('_'*20)

#striping
print('     monika chauhan   '.strip())
print('     monika chauhan   ')
print('_'*20)

#strip replace
print('monika chauhan')
print('monika chauhan'.replace('ka',' '))
print('monika chauhan'.replace('ika','a'))
print('_'*20)

#string count
print('hello guys')
print('hello guys'.count('h'))
print("delhi is the captital of India".count('i'))
print("delhi is the captital of India".lower().count('i'))
print('_'*20)

#string find
print('Mumbai is known as dream city'.find('dream'))
print('_'*20)

#string type check
print("Monika".isalpha())
print('3145'.isdigit())
print('monika'.islower())
print('monika'.isupper())
print('monika'.upper().isupper())
print('_'*20)

#string capitalization
print("monika chauhan".capitalize())
print('monika chauhan'.title())
print('_'*20)

#check for start and end
print("hello everyone".startswith('hello'))
print("hello everyone".startswith('everyone'))
print("hello everyone".endswith('everyone'))
print("hello everyone".endswith('hello'))
print('_'*20)

#
print('monika chauhan'.center(30,'_'))
print('monika chauhan'.ljust(30,'_'))
print('monika chauhan'.rjust(30,'_'))
print('_'*20)


```
