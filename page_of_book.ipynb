```python
import pandas as pd
import requests
from tqdm import tqdm 
from bs4 import BeautifulSoup
df=pd.read_csv("book_scrapping.csv")
df.head()
```

```python
data=[]
for link in tqdm(df["book_link"]):
    res=requests.get(link)
    soup=BeautifulSoup(res.text,"html.parser")
    type_=soup.find("ul",class_="breadcrumb").find_all("a")[2].text
    upc            =soup.find("table",class_="table table-striped").find_all("td")[0].text
    price_x        =soup.find("table",class_="table table-striped").find_all("td")[2].text[1:]
    price_i        =soup.find("table",class_="table table-striped").find_all("td")[3].text[1:]
    tax            =soup.find("table",class_="table table-striped").find_all("td")[4].text[1:]
    quantity       =soup.find("table",class_="table table-striped").find_all("td")[5].text
    review         =soup.find("table",class_="table table-striped").find_all("td")[6]
    data.append([type_,price_x,price_i,tax,upc,review,quantity])
  
```

```python
df1=pd.DataFrame(data,columns=["Category","price_e_tax","price_i_tax","tax","upc","review","qunatity"])
df1.tail()
# df1.to_csv("page_scrapping.csv",index=False)
```
